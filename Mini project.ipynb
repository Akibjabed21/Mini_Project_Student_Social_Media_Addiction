# 2. Exploratory Data Analysis (EDA) 
    ● Understand relationships between:
         ○ Age, Gender, Daily Usage (in hours) 
         ○ Sleep patterns, Academic performance, Social interaction

```python
df.groupby(['Age','Gender'])['Avg_Daily_Usage_Hours'].mean()
```

```python
# Create the boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x='Age', y='Avg_Daily_Usage_Hours', hue='Gender', data=df, palette='Set1')

# Customize the plot
plt.title('Daily Social Media Usage by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Average Daily Usage (Hours)')
plt.legend(title='Gender')
plt.show()
```

*Insight: Younger age groups (e.g., 13–18) show significantly higher usage, especially among females.

### Sleep vs. Usage — Scatter Plot with Trendline

```python
df[['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night']]
```

```python
sns.lmplot(x='Avg_Daily_Usage_Hours', y='Sleep_Hours_Per_Night', data=df, hue='Gender')
plt.title('Sleep Duration vs. Daily Social Media Usage')
plt.xlabel('Daily Usage (Hours)')
plt.ylabel('Sleep Hours')
plt.show()
```

*Insight: Students using social media more than 4 hours/day tend to sleep less than 6 hours.

### Academic Performance vs. Usage — Box Plot

```python
df['Academic_Level'] = pd.cut(df['Avg_Daily_Usage_Hours'], bins=[0,2,4,6,24], labels=['Low','Medium','High','Extreme'])

sns.boxplot(x='Avg_Daily_Usage_Hours', y='Academic_Level', data=df)
plt.title('Academic Performance by Usage Level')
plt.xlabel('Usage Level')
plt.ylabel('Performance Score')
plt.show()
```

*Insight: Students in the “Extreme” usage category show wider performance variability and lower median scores.

### Social Interaction vs. Usage — Heatmap

```python
correlation = df[['Avg_Daily_Usage_Hours', 'Conflicts_Over_Social_Media']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Usage and Social Interaction')
plt.show()
```

*Insight: Negative correlation suggests that more time online may reduce offline social engagement.

# 4. Functions, Loops, and Conditionals 
            ● Create custom functions to: 
            ○ Classify risk level (Low/Medium/High) based on usage hours 
            ○ Suggest digital detox strategies using if-else blocks

```python
# Classify risk level (Low/Medium/High) based on usage hours 
def classify_risk(hours):
    if hours <= 2:
        return "Low"
    elif 2 < hours <= 4:
        return "Medium"
    else:
        return "High"

df["Risk_Level"] = df["Avg_Daily_Usage_Hours"].apply(classify_risk)


print(df[["Student_ID", "Avg_Daily_Usage_Hours", "Risk_Level"]].head(10))
```

```python
def suggest_strategy(risk):
    if risk == 'Low':
        return 'Maintain balance, schedule short breaks.'
    elif risk == 'Medium':
        return 'Limit screen time, use app blockers.'
    else:
        return 'Digital detox plan, counseling support, offline hobbies.'
```

```python
df['Detox_Strategy'] = df['Risk_Level'].apply(suggest_strategy)

print(df[["Student_ID", "Avg_Daily_Usage_Hours", "Detox_Strategy"]].head(10))
```

```python
#Loop-Based Summary:
for index, row in df.iterrows():
    print(f"{row['Student_ID']} is at {row['Risk_Level']} risk. Suggested strategy: {row['Detox_Strategy']}")
```

```python
#Bar Chart 

plt.figure(figsize=(8,4))
sns.barplot(x="Gender", y="Avg_Daily_Usage_Hours", data=df, estimator="mean", errorbar=None)
plt.title("Average Daily Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Avg Daily Usage (hours)")
plt.show()

print(" Insight: On average, one gender tends to spend slightly more time on social media than the other.")
```

```python
#Pie Chart 

plt.figure(figsize=(10,4))

df["Most_Used_Platform"].value_counts().plot.pie(autopct="%1.1f%%",startangle=10,shadow=True, cmap="Set3")
plt.title("Distribution of Most Used Social Media Platforms")
plt.ylabel("")  # Remove y-label
plt.legend(loc='upper right')
plt.axis('equal')
plt.show()

print(" Insight: A few platforms dominate students' social media usage habits.")
```

```python
# Lineplot

plt.figure(figsize=(8,4))
sns.lineplot(x="Avg_Daily_Usage_Hours", y="Sleep_Hours_Per_Night", data=df, marker="o")
plt.title("Relationship between Usage Hours and Sleep Duration")
plt.xlabel("Avg Daily Usage (hours)")
plt.ylabel("Sleep Hours per Night")
plt.show()

print(" Insight: As daily usage increases, average sleep hours tend to decline.")
```

```python
# Hitmsp 

plt.figure(figsize=(8,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis", fmt=".2f")
plt.title("Correlation Heatmap of Key Factors")
plt.show()

print(" Insight: Stronger correlations are observed between daily usage, addicted score, and mental health.")
```
